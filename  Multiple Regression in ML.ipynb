```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the Dataset
data=pd.read_csv("Toyota.csv", na_values=["??"])
data=data.dropna()
data
```

```python
# Split the data based on independent (x) and dependent variable (y)
x=data.iloc[:,1:3].values #x-features (Age,KM)
y=data.iloc[:,0:1].values #y-labels (price)
```

```python
# Divide the complete dataset into training and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#print(x_train)
```

```python
# To create my model
# Regression Line on Train dataset
from sklearn.linear_model import LinearRegression
lrc=LinearRegression() #object
lrc.fit(x_train,y_train) #Training
```

```python
print(lrc.coef_) #label(Predict)=intercept_ + coef1_ * 'Age' + coef2_ * 'KM'
print(lrc.intercept_)
```

```python
predict=lrc.predict(x_test)
print(predict) #Predicted values of Price
print(y_test) #Actual Values of Price
```

```python
plt.scatter(y_test,predict)
plt.show()
```

```python
plt.plot(y_test,predict,'r+')
```

```python
#Find Accuracy of data

print("Accuracy of Fitting the Model:",lrc.score(x_test,predict))

print("Accuracy of Test Data:",lrc.score(x_test,y_test))

print("Accuracy of Train Data:",lrc.score(x_train,y_train))
```
