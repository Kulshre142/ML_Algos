```python
#binary step activation function
import numpy as np
import matplotlib.pyplot as plt

def binaryStep(x):
    return np.heaviside(x,1)

x = np.linspace(-10, 10) #List
plt.plot(x, binaryStep(x))
#plt.plot(x, binaryStep(x),'r+')
plt.title('Activation Function :binaryStep')
plt.show()
```

```python
#linear activation function
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    return x

x = np.linspace(-10, 10)
#plt.plot(x, linear(x))
plt.plot(x, linear(x),'r+')
plt.title('Activation Function :Linear')
plt.show()
```

```python
#sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
#Tanh activation function
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x),'r+')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
#ReLU Aactivation function
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.title('Activation Function :RELU')
plt.show()
```

```python
#softmax function
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.title('Activation Function :Softmax')
plt.show()
```
